4.Analyzing each image

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
```

5. Listing all tif image in the directory

```python
image_files = [f for f in os.listdir() if f.lower().endswith('.tif')]
print("Total images found:",len(image_files))
print("Examples files:",image_files[:5])
```

6. Analyzing all the Images with different parameteres

```python
image_data = []

for img_file in image_files:
    try:
        with Image.open(img_file) as img:
            width, height = img.size
            mode = img.mode
            fmt = img.format
            dpi = img.info.get('dpi', (0, 0))
            file_size_kb = os.path.getsize(img_file) / 1024
            aspect_ratio = round(width / height, 2)
            is_square = width == height

            image_data.append({
                'filename': img_file,
                'width': width,
                'height': height,
                'format': fmt,
                'mode': mode,
                'file_size_kb': round(file_size_kb, 2),
                'aspect_ratio': aspect_ratio,
                'is_square': is_square,
                'dpi_x': dpi[0],
                'dpi_y': dpi[1]
            })
    except Exception as e:
        print(f"Error with file {img_file}: {e}")
```

7. Showing images

```python
import matplotlib.pyplot as plt

for i in range(min(5, len(image_files))):
    img = Image.open(image_files[i])
    plt.imshow(img, cmap='gray')
    plt.title(image_files[i])
    plt.axis('off')
    plt.show()
```

8. Saving The Analysis image data to CSV

```python
# Converting to DataFrame
df = pd.DataFrame(image_data)

# Saving CSV file
df.to_csv("hyphal_image_analysis.csv", index=False)

# Downloading CSV file
from google.colab import files
files.download("hyphal_image_analysis.csv")
```
